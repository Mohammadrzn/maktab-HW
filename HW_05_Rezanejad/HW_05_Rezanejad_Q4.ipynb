```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("iris_dataset.csv")
df
```

```python
groups = df.groupby("target")
```

```python
means = groups.mean()
means
```

```python
standard_deviation = groups.std()
standard_deviation
```

```python
ranges = groups.max() - groups.min()
ranges
```

```python
setosa = df[df["target"] == 0]
setosa_df = pd.DataFrame({
    "Mean": setosa.mean(),
    "Standard deviation": setosa.std(),
    "Range": setosa.max() - setosa.min()
})
setosa_df
```

```python
versicolor = df[df["target"] == 1]
versicolor_df = pd.DataFrame({
    "Mean": versicolor.mean(),
    "Standard deviation": versicolor.std(),
    "Range": versicolor.max() - versicolor.min(),
})
versicolor_df
```

```python
virginica = df[df["target"] == 2]
virginica_df = pd.DataFrame({
    "Mean": virginica.mean(),
    "Standard deviation": virginica.std(),
    "Range": virginica.max() - virginica.min(),
})
virginica_df
```

```python
summary = pd.concat(
    {
        "Mean": means,
        "Standard deviation": standard_deviation,
        "Range": ranges
    },
    axis=1
)

summary
```

```python
setosa_correlation_matrix = setosa_df.corr()
plt.figure(figsize=(8,6))
sns.heatmap(setosa_correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Setosa Correlation Heatmap")
plt.show()
```

```python
versicolor_correlation_matrix = versicolor_df.corr()
plt.figure(figsize=(8,6))
sns.heatmap(versicolor_correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Versicolor Correlation Heatmap")
plt.show()
```

```python
virginica_correlation_matrix = virginica_df.corr()
plt.figure(figsize=(8,6))
sns.heatmap(virginica_correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Virginica Correlation Heatmap")
plt.show()
```
