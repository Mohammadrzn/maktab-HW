```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("iris_dataset.csv")
df
```

```python
x = df[["sepal length (cm)", "sepal width (cm)"]]
y = df["target"]
```

```python
p_test = x.iloc[-1]
real_class = y.iloc[-1]
x_train = x.iloc[:-1]
y_train = y.iloc[:-1]
euclidean_distance = np.sqrt((x_train["sepal length (cm)"] - p_test["sepal length (cm)"])**2 + (x_train["sepal width (cm)"] - p_test["sepal width (cm)"])**2)
```

```python
idx = euclidean_distance.idxmin()
nearest_class = y_train.loc[idx]
```

```python
print("Predicted:", nearest_class)
print("Real:", real_class)
```

```python
k = 3
k_idx = euclidean_distance.nsmallest(k).index
k_classes = y_train.loc[k_idx]
predicted_k3 = k_classes.mode()[0]
print("predicted_k3:", predicted_k3)
```
