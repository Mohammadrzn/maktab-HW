```python
import matplotlib.pyplot as plt
import numpy as np

ref = plt.imread("face_reference.jpg")
inp = plt.imread("face_input.jpg")
```

```python
def to_gray(img):
    if len(img.shape) == 3:
        return img.mean(axis=2)
    return img

ref_gray = to_gray(ref)
inp_gray = to_gray(inp)
```

```python
min_shape = (
    min(ref_gray.shape[0], inp_gray.shape[0]),
    min(ref_gray.shape[1], inp_gray.shape[1])
)

ref_gray = ref_gray[:min_shape[0], :min_shape[1]]
inp_gray = inp_gray[:min_shape[0], :min_shape[1]]
```

```python
MAE = np.mean(np.abs(ref_gray - inp_gray))
MSE = np.mean((ref_gray - inp_gray)**2)

MAE, MSE
```
